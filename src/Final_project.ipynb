### Check for missing values

```python
print(amazon_df.isna().sum())
```

As you can see there're 2 missing values in attribute `rating_count`, so i'm gonna remove it.

```python
amazon_df.dropna(subset=['rating_count'], inplace=True)
print(amazon_df.isna().sum())
```

Changing the data type of discounted price, actual price and discount percentage

```python
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

```python
amazon_df.head()
```

```python
amazon_df.info()
```

There's something weird about attribute `rating`, i'll show you

```python
amazon_df['rating'].value_counts()
```

There's a row in dataset contains a special character `|`, i will exclude it

```python
amazon_df = amazon_df[amazon_df['rating'].apply(lambda x: '|' not in str(x))]
amazon_df.head()
```

```python
amazon_df['rating'].value_counts()
```

Changing the data type of rating count and rating

```python
amazon_df['rating'] = amazon_df['rating'].astype('float64')
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(',', '').astype('float64')
amazon_df.info()
```

### Describe the correlation between numeric features in the dataset
