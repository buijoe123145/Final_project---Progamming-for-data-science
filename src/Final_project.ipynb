**Question 1:** Based on the quantity of products, what can be inferred from the top 30 primary categories and top 30 subcategories?

```python
sub_category_counts = amazon_df['sub_category'].value_counts()[:30]
categories = sub_category_counts.index
counts = sub_category_counts.values

colors = sns.color_palette('pastel', len(categories))
plt.figure(figsize=(15, 10))
bar_plot =sns.barplot(x=categories, y=counts, palette=colors)
plt.ylabel('Counts',fontweight='bold')
plt.xlabel(None)
plt.title('Number of Products per Sub Category',fontweight='bold')
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.legend(bar_plot.patches, categories, title='Sub Categories', loc='upper right')
plt.show()
```

### Question 4: Which categories are most often discounted with high discount percentage(>50%) ?

Meaning: Customer can wait to sale event to buy their wishlist at the lowest possible price if the product is in the category of regular discounts?

```python
# choose necessary columns to answer this question
regularBigSale = amazon_df.copy()[['category','discount_percentage']]

regularBigSale.head()
```

```python


# Tách mỗi category thành một dòng riêng biệt
categories_expanded = regularBigSale['category'].str.split('|', expand=True).stack()

# Kết hợp với DataFrame gốc và đặt lại index
result_df = (categories_expanded
             .reset_index(level=1, drop=True)
             .rename('category')
             .to_frame()
             .join(regularBigSale[['discount_percentage']])
             .reset_index(drop=True))
result_df
```

```python
# Tính mean cho mỗi category và sắp xếp theo số lượng giảm dần
category_means = result_df.groupby('category')['discount_percentage'].mean().reset_index()
category_counts = result_df['category'].value_counts().sort_values(ascending=False)

# Lọc các category có mean >= 0.5
filtered_categories = category_means[category_means['discount_percentage'] >= 0.5]['category']

# Lấy top 30 category theo số lần giảm giá nhiều nhất
top_categories = category_counts[filtered_categories].head(30).index

# Lọc dữ liệu theo các category này
filtered_data = result_df[result_df['category'].isin(top_categories)]
```

```python
# Vẽ boxplot
plt.figure(figsize=(20, 10))
sns.boxplot(y='category', x='discount_percentage', data=filtered_data.explode('discount_percentage'), palette='turbo')
plt.title('Boxplot of Discount Percentage by Category (Mean >= 0.5, Top 30 Categories)')
plt.xlabel('Discount Percentage')
plt.ylabel('Category')

plt.axvline(x=0.5, color='red', linestyle='--', label='Threshold at 0.5')
plt.show()
```
