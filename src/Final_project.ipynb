### Check for missing values

```python
print(amazon_df.isna().sum())
```

As you can see there're 2 missing values in attribute `rating_count`, so i'm gonna remove it.

```python
amazon_df.dropna(subset=['rating_count'], inplace=True)
print(amazon_df.isna().sum())
```

Changing the data type of discounted price, actual price and discount percentage

```python
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%','').astype(float)/100
```

```python
amazon_df.head()
```

```python
amazon_df.info()
```

There's something weird about attribute `rating`, i'll show you

```python
amazon_df['rating'].value_counts()
```

There's a row in dataset contains a special character `|`, i will exclude it

```python
amazon_df = amazon_df[amazon_df['rating'].apply(lambda x: '|' not in str(x))]
amazon_df.head()
```

```python
amazon_df['rating'].value_counts()
```

Changing the data type of rating count and rating

```python
amazon_df['rating'] = amazon_df['rating'].astype('float64')
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(',', '').astype('float64')
amazon_df.info()
```

Categories on the e-commerce site **Amazon** are organized in order of relevance; categories with more relevance to the product are placed first, and vice versa. So, I add two new columns, one labeled `main_category` for the category in the first position and another labeled `sub_category` for the category in the last position.

```python
amazon_df['main_category'] = amazon_df['category'].str.split('|').str[0]
amazon_df['sub_category'] = amazon_df['category'].str.split('|').str[-1]
amazon_df.head()
```

## Asking/Answering question

**Question 1:** Based on the quantity of products, what can be inferred from the top 30 primary categories and top 30 subcategories?

```python
main_category_counts = amazon_df['main_category'].value_counts()[:30]
categories = main_category_counts.index
counts = main_category_counts.values

plt.figure(figsize=(15, 8))
sns.barplot(x=categories, y=counts, palette='turbo')
plt.xticks(rotation=30)
plt.ylabel('Counts',fontweight='bold')
plt.xlabel(None)
plt.title('Number of Products per Main Category',fontweight='bold')
plt.show()
```

- Electronics, Computers & Accessories, and Home & Kitchen are the top three key categories. This indicates that consumers like these categories.The relatively small quantity of goods in the other three primary categories suggests that these categories are not as well-liked as the top three.
-  Office Products, Musical Instruments, Home Improvement, Toys & Games, Car & Motorbike, and Health & Personal Care have a very small number of products, which may suggest that these categories have less demand.
- By analyzing the data, companies may get insight into the state of the industry and spot growth prospects in certain product categories.


```python
sub_category_counts = amazon_df['sub_category'].value_counts()[:30]
categories = sub_category_counts.index
counts = sub_category_counts.values

colors = sns.color_palette('pastel', len(categories))
plt.figure(figsize=(15, 10))
bar_plot =sns.barplot(x=categories, y=counts, palette=colors)
plt.ylabel('Counts',fontweight='bold')
plt.xlabel(None)
plt.title('Number of Products per Sub Category',fontweight='bold')
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.legend(bar_plot.patches, categories, title='Sub Categories', loc='upper right')
plt.show()
```

- Smartwatches, smartphones, smart TVs, in-ear headphones, USB cables, and remote controllers make up the top six category. These are the most well-liked subcategories, therefore companies should concentrate on offering goods in these areas to get clients.
- The information reveals that the top 30 categories include a wide variety of subcategories, such as home electronics, personal items, and kitchen equipment. This emphasizes how crucial it is to provide a range of items in order to meet the various demands and tastes of customers.
